```python
# Plot the data as a bar plot
plt.figure(figsize=(12, 8))  # Set the figure size

# Create a bar plot (for the top 10 states for example)
plt.bar(tobacco_by_state_sorted['LocationDesc'][:10], tobacco_by_state_sorted['DataValue'][:10], color='skyblue')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Total Tobacco Use')
plt.title('Top 10 States by Total Tobacco Use')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

```python
import geopandas as gpd
import folium

# Load the downloaded shapefile (adjust the path to where you extracted the files)
shapefile_path = r'C:\Users\aysen\OneDrive\Desktop\project-3/ne_110m_admin_0_countries.shp'

# Read the shapefile using geopandas
world = gpd.read_file(shapefile_path)


# Filter for only the United States (based on the "NAME" column)
us = world[world['NAME'] == 'United States of America']

# Initialize a folium map centered on the US
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Use folium's built-in Choropleth method
folium.Choropleth(
    geo_data=us.__geo_interface__,  # GeoJSON representation of the GeoDataFrame
    data=tobacco_by_state_sorted,  # Your tobacco use data
    columns=['LocationDesc', 'DataValue'],  # Data columns: state and values
    key_on='properties.NAME',  # Match the GeoJSON field with the state names
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Tobacco Use'
).add_to(m)

# Add layer control
folium.LayerControl().add_to(m)

# Save the map as HTML
m.save('tobacco_use_by_state_map.html')

# Display the map in Jupyter (if using notebook)
m
```
