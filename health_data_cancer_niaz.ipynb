```python
# Filter based on the specified conditions and exclude "Invasive cancer (all sites combined), incidence"
filtered_df = health_data_df[
    (health_data_df['DataSource'] == 'Statewide central cancer registries') &
    (health_data_df['Question'] != 'Invasive cancer (all sites combined), incidence')
]

# Group by the filtered DataFrame and calculate the mean of DataValue
state_mean_data_value = filtered_df.groupby(['Question'])['DataValue'].mean().reset_index()

# Sort by DataValue in descending order
state_mean_data_value_sorted = state_mean_data_value.sort_values(by='DataValue', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
plt.barh(state_mean_data_value_sorted['Question'], state_mean_data_value_sorted['DataValue'], color='red')

# Adding labels and title
plt.xlabel('Avg Annual Per 100,000 People')
plt.ylabel('Specific Cancer Type')
plt.title('Cancer in US Population')

plt.tight_layout()  # Adjusts layout to fit labels and title properly
plt.show()
```
